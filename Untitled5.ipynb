```python
from lxml.html import fromstring


def get_proxies():
    url = 'https://free-proxy-list.net/'
    response = requests.get(url)
    parser = fromstring(response.text)
    proxies = set()
    for i in parser.xpath('//tbody/tr')[:20]:
        if i.xpath('.//td[7][contains(text(),"yes")]'):
            proxy = ":".join([i.xpath('.//td[1]/text()')[0], i.xpath('.//td[2]/text()')[0]])
            proxies.add(proxy)
    return proxies
```

```python
def scrape_etsy(keywords: list, items_to_scrape: int) -> pd.DataFrame:
    proxies = get_proxies()
    average_items_per_page = 64
    pages_to_scrape = math.ceil(items_to_scrape / average_items_per_page)
    df_list = []

    for key in keywords:
        titles, prices, item_urls, img_urls = ([] for i in range(4))
        adj_keyword = key.lower()

        for page_no in range(1, pages_to_scrape + 1):
            for proxy in proxies:
                try:
                    page = requests.get(
                        f"https://www.aruodas.lt/{adj_keyword}/puslapis/{page_no}/",
                        headers = {
                            'authority': 'www.etsy.com',
                            'sec-ch-ua': '^\\^Google',
                            'sec-ch-ua-mobile': '?0',
                            'upgrade-insecure-requests': '1',
                            'user-agent': f'{ua.random}',
                            'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
                            'sec-fetch-site': 'same-origin',
                            'sec-fetch-mode': 'navigate',
                            'sec-fetch-user': '?1',
                            'sec-fetch-dest': 'document',
                            'referer': 'https://www.etsy.com/', # add changing referer
                            'accept-language': 'en-US,en;q=0.9',
                        },
                        proxies={"http": str(proxy), "https": str(proxy)},
                        params = (
                            ('explicit', '1'),
                            ('ref', 'pagination'),
                            ('page', f'{page_no}'),
                        )
                        {'FOrder': 'AddDate'} # ADD IF WANT NEWEST FIRST
                    )
                except requests.exceptions.RequestException as e:
                    continue
                else:
                    break

            soup = BeautifulSoup(page.content, 'html.parser')

            for value in soup.find_all('div', attrs={'class': "list-photo"}):
                  titles.extend([title.img['alt'] for title in value.find_all('a')])
                  item_urls.extend([link['href'] for link in value.find_all('a')])
                  img_urls.extend([title.img['src'] for title in value.find_all('a')])

            for price in soup.find_all('span', attrs={'class': 'list-item-price'}):
                    converted_price = "".join(re.findall(r'\d+', str(price)))
                    prices.append(int(converted_price))

        df_list.append(pd.DataFrame({
        'category': adj_keyword,
        'title': titles,
        'price': prices,
        'item_url': item_urls,
        'img_url': img_urls
    }))
   
    return pd.concat(df_list, ignore_index=True)
```
